### creating a function to clean phone number

```python
import re
def clean_phone(phone):
    phone = str(phone).split('x')[0]   # Remove extension
    digits = re.sub(r'\D', '', phone)
    return int(digits)

data['Phone'] = data['Phone'].apply(clean_phone)
```
